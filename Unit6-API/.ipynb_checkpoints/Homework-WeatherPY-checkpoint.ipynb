## Analysis

#### 1. The max temperature peaks near the equator. Temperature Decreases as we move away from the equator.
#### 2. Humidity and cloudiness are all over the place and do not seem to be correlated to latitude.
#### 3. The majority of cities have a windspeed under 20 mph and it seems to increase as we move away from the equator

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from  config_weather import api_key

# Incorporated citipy to determine city based on latitude and longitude

#!pip install citipy

from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial" 
```

```python
# Create empty lists to append 
City = []
Country = []
Temperature = []
Humidity = []
Cloudiness = []
Wind_speed = []
Date = []
Latitude = []
Longitude = []



# Start the counter 
city_num = 1

# Log 
print(f"Start Data Extract")
print(f"-------------------------------")

#Loop through the cities in the list 
for city in cities:
    try: 
        response = requests.get(f"{url}&appid={api_key}&q={city}").json() 
        City.append(response["name"])
        Cloudiness.append(response["clouds"]["all"])
        Country.append(response["sys"]["country"])
        Date.append(response["dt"])
        Humidity.append(response["main"]["humidity"])
        Temperature.append(response["main"]["temp_max"])
        Latitude.append(response["coord"]["lat"])
        Longitude.append(response["coord"]["lon"])
        Wind_speed.append(response["wind"]["speed"])
        city_name = response["name"]
        country_name = response['sys']['country']
        print(f"Processing City {city_num} | {city_name} | {country_name}")
        print(f"{url}&q={city}")
        
        # Increase counter by one 
        city_num= city_num + 1
        
        # Wait 
        time.sleep(1.10)
        
    # If no record found "skip" to next call
    except:
        print("City not found. Skip!")
    continue

print(f"-------------------------------")
print(f"-------------------------------")
print(f"-------------------------------")
print(f"Data Extract Completed!")
```

```python
#Count of each list

print(f"City count {len(City)}")
print(f"Country count {len(Country)}")
print(f"Cloudiness count {len(Cloudiness)}")
print(f"Humidity count {len(Humidity)}")
print(f"Latitude count {len(Latitude)}")
print(f"Longitude count {len(Longitude)}")
print(f"Temperature count {len(Temperature)}")
print(f"Wind speed count {len(Wind_speed)}")
print(f"Date count {len(Date)}")


```

```python
# Create a data frame from generated lists
weather_df = pd.DataFrame({
    "City": City,
    "Country": Country,
    "Cloudiness":Cloudiness, 
    "Humidity": Humidity,
    "Latitude":Latitude, 
    "Longitude":Longitude, 
    "Temperature": Temperature,
    "Wind Speed":Wind_speed,
    "Date":Date
}
)

# Display count of weather data values 
weather_df.count()
```

```python
# Save data frame to CSV
weather_df.to_csv('weather_data.csv')

# Display data frame  
weather_df.head()

```

```python
# Changing Date Format
from datetime import datetime as dt

weather_df['Date'] = dt.now().strftime("%m/%d/%Y") 

weather_df.head()
```

```python
# Build a scatter plot for each data type
plt.figure(figsize=(20,10))
           
plt.scatter(weather_df["Latitude"], weather_df["Temperature"], marker="o", s=30, color = 'blue')

# Incorporate the other graph properties
plt.title("Latitude vs. Max Temperature",fontsize=20)
plt.ylabel("Max. Temperature (F)",fontsize=15)
plt.xlabel("Latitude",fontsize=15)
plt.grid(True)

# Save the figure
plt.savefig("Max_Temp_vs_Latitude.png")

# Show plot
plt.show()
```
